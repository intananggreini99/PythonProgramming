```python
def intRangeCek(number,numMin,numMax):
  while True:
    try:
        numRange = [x for x in range(int(numMin),int(numMax))]

        if not number.is_integer():
          raise TypeError

        if number not in numRange:
            raise ValueError
        else:
          print(f"The number is: {int(number)}")
          break

    except TypeError:
      print("Error: wrong input")
      number = float(input("Enter a number from -10 to 10: "))
    except ValueError:
      print(f"Error: the value is not within permitted range ({int(numMin)}..{int(numMax)})")
      number = float(input("Enter a number from -10 to 10: "))

inNumMin = int(input("Enter a number for low acceptable limit: "))
inNumMax = int(input("Enter a number for high acceptable limit: "))
inNum = float(input("Enter a number from -10 to 10: "))
intRangeCek(inNum,inNumMin,inNumMax)
```

```python
import re

def cariSomething():
    try:
        file_path = input("Masukkan nama file: ")
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()

        pola = input("Masukkan pola regex: ")
        cocokin = re.findall(pola, text, flags=re.IGNORECASE)
        try:
          assert cocokin, "Tidak ada data yang ditemukan!"  # Akan raise AssertionError jika cocokin kosong
          for objek in cocokin:
              print(f"Data ditemukan: {objek}")
        except AssertionError as a:
            print(a)

    except FileNotFoundError:
        print(f"Error: File {file_path} tidak ditemukan!")
    except PermissionError:
        print("Error: Izin file ditolak!")
    except KeyboardInterrupt:
        print(f"Error: User lelah menginput data.")
    else :
      print("Senang bisa membantu :)")
    finally:
      print("Program selesai.")

print("Isi /content/sample_data/:")
!ls /content/sample_data
cariSomething()
```

# **Exercise 2** : Catatan Belanja Pribadi


```python
belanjaan = {}

def belanja_bulanan(dana, belanjaan):
    try:
        total = 0
        print(f"\n")
        for item, harga in belanjaan.items():
            total += harga
            print(f" {item}: Rp{harga:,}")

        print(f"\nDana yang dibutuhkan: Rp{total:,}")

        print(f"Dana tersedia: Rp{dana:,}")
        if dana >= total :
          print("Dana yang tersedia mencukupi untuk belanja.")
          print(f" Belanja selesai! Sisa dana: Rp{dana - total:,}")
        else:
          print(f"Total belanja Rp{total:,} melebihi dana")
          raise ValueError("Dana tidak cukup untuk belanja bulanan")

    except ValueError as e:
        print(f" Tidak bisa belanja: {e}")
    else:
        print(" Pengeluaran hari ini tercatat")
    finally:
        print("Save Your Money :)")


print("Masukkan belanjaan (ketik 'selesai' untuk berhenti)")
while True:
    produk = input("Nama produk: ")
    if produk.lower() == 'selesai':
        break

    try:
        harga = int(input("Harga: Rp"))
        belanjaan[produk] = harga
    except ValueError:
        print("Harga harus angka! Coba lagi")

danaTersedia = int(input(f"Masukkan jumlah dana yang tersedia: Rp"))
belanja_bulanan(danaTersedia, belanjaan)
```
